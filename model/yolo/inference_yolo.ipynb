```python
from ultralytics import YOLO
from tqdm import tqdm


model = YOLO(f"yolo11n.yaml")
model = YOLO("/data/ephemeral/home/level2-objectdetection-cv-20/code/yolo/runs/detect/n_40_b_489/weights/best.pt")  # pretrained YOLO11n model
source = "../../dataset/test/"

# Run batched inference on a list of images
results = model(source, imgsz=1024)  # return a list of Results objects
```

```python
import pandas as pd
from pycocotools.coco import COCO

prediction_strings = []
file_names = []
annotation = '../../dataset/test.json'
coco = COCO(annotation)

# submission 파일 생성
for i, output in enumerate(tqdm(results)):
    prediction_string = ''
    image_info = coco.loadImgs(coco.getImgIds(imgIds=i))[0]
    for box in output.boxes:
        if box.conf > 0.05: 
            prediction_string += str(int(box.cls)) + ' ' + str(float(box.conf)) + ' ' + str(float(box.xyxy[0][0])) + ' '  + str(float(box.xyxy[0][1])) + ' ' + str(float(box.xyxy[0][2])) + ' ' + str(float(box.xyxy[0][3])) + ' '
    prediction_strings.append(prediction_string)    
    file_names.append(image_info['file_name'])
    # file_names.append(f"test/{i:04}.jpg")



submission = pd.DataFrame()
submission['PredictionString'] = prediction_strings
submission['image_id'] = file_names
submission.to_csv('./test_ni_submission_005.csv', index=None)
```
